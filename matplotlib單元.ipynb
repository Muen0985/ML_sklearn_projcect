```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(8) # 0~7
y=x*2
```

```python
a=np.linspace(0,10,11)
b=a**2
```

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0,7)
plt.ylim(0,15)
plt.title('my_plot1')
plt.show()
```

```python
fig=plt.figure(figsize=(5,3))  # 如果可以用figsize 可能不用額外設定axes
ax=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.1,0.35,0.15,0.15])

ax.plot(a,b)
ax.set_xlim(0,11)
ax.set_ylim(0,110)
ax.set_xlabel("time")
ax.set_ylabel('number')
ax.set_title('Number of branches')

ax2.plot(a,b)
ax2.set_xlim(5,10)
ax2.set_ylim(30,70)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Zoomed In');
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(6,4))
ax[0,0].plot(x,y)  # ax是一個numpy nd.array
ax[0,0].set_ylabel('gallon')
ax[1,1].plot(a,b)
ax[1,1].set_title('hey')

# plt.tight_layout() # 讓圖不要有重疊，自動調整

fig.subplots_adjust(wspace=0.2,hspace=0.35) #手動調整

fig.suptitle('Test plots',fontsize=16) # super title
plt.show()
```

```python
fig= plt.figure(figsize=(5,3))
ax= fig.add_axes([0,0,1,1])

ax.plot(x,x,label='just x',color='purple',lw=2,marker='o')
ax.plot(a,b,label='exponential a',ls='--')

ax.legend(loc=(1.1,0.7)) # loc=0 best 
# 也可以給定x,y值，放在圖外面
plt.show()
```
