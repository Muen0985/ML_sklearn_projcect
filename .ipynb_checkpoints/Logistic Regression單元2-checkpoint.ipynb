```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('UNZIP_FOR_NOTEBOOKS_FINAL/DATA/iris.csv')
```

```python
df.head()
```

```python
df['species'].value_counts()
```

```python
plt.figure(figsize=(3,2))
sns.countplot(x='species',data=df)
plt.show()
```

```python
x=df.drop('species',axis=1)
y=df['species']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=101)
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```python
log_model=LogisticRegression(solver='saga',multi_class='ovr',max_iter=5000)
```

```python
penalty=['l1','l2','elasticnet']
l1_ratio=np.linspace(0,1,10)
C=np.logspace(0,5,10)
param_grid={
    'penalty':penalty,
    'l1_ratio':l1_ratio,
    'C':C
}
```

```python
gridmodel=GridSearchCV(log_model,param_grid=param_grid)
```

```python
gridmodel.fit(x_train,y_train) # 有warnings是正常的，因為只有penalty是elasticnet時，才會有l1_ratio這個參數
```

```python
gridmodel.best_params_
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,plot_confusion_matrix
```

```python
y_pred=gridmodel.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
plot_confusion_matrix(gridmodel,x_test,y_test)
plt.show()
```

```python
print(classification_report(y_test,y_pred))
```
