```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# using the mpimg to convert .jpg .png into numpy array
```

```python
img_arr = mpimg.imread('UNZIP_FOR_NOTEBOOKS_FINAL/DATA/palm_trees.jpg')
```

```python
img_arr.shape  # H ,W ,C
```

```python
plt.imshow(img_arr)  # needs the numpy array to show
plt.show()
```

```python
(h,w,c)=img_arr.shape
```

```python
print(h, w, c)
```

```python
img_arr2d=img_arr.reshape(h*w,c)
```

```python
len(img_arr2d.shape) # 2 dimensional array
```

```python
from sklearn.cluster import KMeans
model= KMeans(n_clusters=6)
labels=model.fit_predict(img_arr2d)
```

```python
rgb_codes=model.cluster_centers_.round(0).astype(int)
```

```python
rgb_codes  # 6 cluster center points
```

```python
labels
```

```python
rgb_codes[labels]  # cluster the point by labels
```

```python
quantized_img = np.reshape(rgb_codes[labels],(h,w,c))
```

```python
plt.imshow(quantized_img)
plt.show()
```
