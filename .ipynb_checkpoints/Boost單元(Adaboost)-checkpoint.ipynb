# Modeling

```python
x=df.drop('class',axis=1)
x=pd.get_dummies(x,drop_first=True)
y=df['class']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.15, random_state=101)
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
model=AdaBoostClassifier(n_estimators=1) 
# how many stumps we want to create 這裡的意思是如果要從資料裡只挑出一個特徵來辨認的話，模型會選哪一個
```

```python
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import classification_report,plot_confusion_matrix,accuracy_score
```

```python
predictions = model.predict(X_test)
```

```python
print(classification_report(y_test,predictions))
```

```python
model.feature_importances_ # 模型所選擇的最重要項目(一項)
```

```python
model.feature_importances_.argmax()  #返回的是列表中最大值的位置
```

```python
error_rates=[]

for n in range(1,96):
    model=AdaBoostClassifier(n_estimators=n)
    model.fit(X_train,y_train)
    preds=model.predict(X_test)
    
    error=1-accuracy_score(y_test,preds)
    error_rates.append(error)
```

```python
plt.figure(figsize=(5,3))
plt.plot(range(1,96),error_rates)
plt.show()
```

```python
model=AdaBoostClassifier(n_estimators=18) # 看起來18是較好的選擇
model.fit(X_train,y_train)
```

```python
feats=pd.DataFrame(index=x.columns,data=model.feature_importances_,columns=['Importance'])
```

```python
imp_feats = feats[feats['Importance']>0]
imp_feats = imp_feats.sort_values("Importance")
```
