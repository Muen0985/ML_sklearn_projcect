```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('UNZIP_FOR_NOTEBOOKS_FINAL/05-Seaborn/dm_office_sales.csv')
df
```

```python
plt.figure(figsize=(10,4))
sns.scatterplot(x='salary',y='sales',data=df, hue='level of education',size='salary',style='level of education') 
# hue:挑選另外一欄位當作分割識別  size:可以依照欄位或單純設定大小 style:設置不同的圖案
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.rugplot(x='salary',data=df)
plt.show()
```

```python
plt.figure(figsize=(6,4))
# sns.set(style='darkgrid')   # 會讓之後所有的output都是這個形式 或是要restart kernel
sns.displot(data=df,x='salary',bins=10,color='blue',kde=True)
```

```python
sns.kdeplot(data=df,x='salary',bw_adjust=0.5) # only show kde plot
```

```python
df['division'].value_counts()
```

```python
df['level of education'].value_counts()
```

```python
plt.figure(figsize=(10,4))
sns.countplot(data=df,x='level of education',hue='division') # countplot的y軸只是去計算數量而已
plt.legend(bbox_to_anchor=(1.05,1))
```

```python
plt.figure(figsize=(10,4))
sns.barplot(data=df,x='level of education',y='salary',estimator=np.mean,ci='sd')
```
