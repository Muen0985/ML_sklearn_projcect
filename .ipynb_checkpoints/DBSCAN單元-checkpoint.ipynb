```python
circles=pd.read_csv('UNZIP_FOR_NOTEBOOKS_FINAL/data/cluster_circles.csv')
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(data=circles,x='X1',y='X2')
plt.show()
```

```python
def display_categories(model,data):
    labels=model.fit_predict(data)
    plt.figure(figsize=(4,3))
    sns.scatterplot(data=data,x='X1',y='X2',hue=labels,palette='Set1')
    plt.show()
```

```python
from sklearn.cluster import KMeans
model=KMeans(n_clusters=2)  # 3 or 2
```

```python
display_categories(model,circles)  # blobs or moons or circles
```

```python
from sklearn.cluster import DBSCAN
```

```python
modeldb=DBSCAN()
```

```python
display_categories(modeldb,blobs)  # blobs or moons or circles
# -1 is identifies as outlier
```
