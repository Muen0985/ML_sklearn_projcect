```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('UNZIP_FOR_NOTEBOOKS_FINAL/DATA/rock_density_xray.csv')
```

```python
df.columns=['signal','density']
```

```python
df.head()
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x='signal',y='density',data=df)
plt.show()
```

```python
# The scale of the data aren't to wide no need to use standardscaler

#  Reshape your data either using array.reshape(-1, 1) if your data has a single feature or,
#  array.reshape(1, -1) if it contains a single sample.

x=df['signal'].values.reshape(-1,1) # to tell sklearn this is a single feature!
y=df['density']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
lr_model=LinearRegression()
lr_model.fit(x_train,y_train)
```

```python
lr_preds=lr_model.predict(x_test)
```

```python
lr_preds # even though the metrics seems good , it report back average and almost the same!
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error

print(mean_absolute_error(y_test,lr_preds))
print(np.sqrt(mean_squared_error(y_test,lr_preds)))
```

```python
arsignal=np.arange(0,100)
```

```python
mysignal_preds=lr_model.predict(arsignal.reshape(-1,1))
```

```python
# mysignal_preds   It's all 2.22...
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='signal',y='density',data=df)
plt.plot(arsignal,mysignal_preds)  # plot out the curve ， turns out it can't fit the data!
plt.show()
```

### Conclusion: So, we need to take a look at the values we predicted versus the true value , can't only rely on  the error metrics

### Use other models

```python
def run_model(model,X_train,y_train,X_test,y_test):
    model.fit(X_train,y_train)
    pred=model.predict(X_test)
    
    print(mean_absolute_error(y_test,pred))
    print(f'rmse: {np.sqrt(mean_squared_error(y_test,pred))}')
    
    signal_range=np.arange(0,100)
    output=model.predict(signal_range.reshape(-1,1))
    
    plt.figure(figsize=(6,4))
    sns.scatterplot(x='signal',y='density',data=df,color='black')
    plt.plot(signal_range,output)
    plt.show()
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
```

```python
pipe=make_pipeline(PolynomialFeatures(degree=5),LinearRegression())  # checkout how to decide the best degree?
```

#### polynomial regression

```python
run_model(pipe,x_train,y_train,x_test,y_test)  
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
tree=[10,20,30,50]
for t in tree:
    print(f'estimators= {t} trees')
    rfr=RandomForestRegressor(n_estimators=t)
    run_model(rfr,x_train,y_train,x_test,y_test)  
```
