```python
df['total_bill'].apply(lambda bill: bill*2)
```

```python
def quality(total_bill,tip):
    if tip/total_bill >0.25:
        return 'generous'
    else:
        return 'others'
```

```python
df[['total_bill','tip']].apply(lambda df: quality(df['total_bill'],df['tip']),axis=1)
# axis=1 針對column做處理 (apply function to each row)
```

```python
df['Quality']=np.vectorize(quality)(df['total_bill'],df['tip']) # 效率較高，推薦這樣做
df
```
