```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('UNZIP_FOR_NOTEBOOKS_FINAL/DATA/cancer_tumor_data_features.csv')
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_x=scaler.fit_transform(df)
```

```python
from sklearn.decomposition import PCA
```

```python
pca_model=PCA(n_components=2)
```

```python
# fit: calculate the eigen values and eigen vectors
# transform: does the project on eigen vectors
pca_model.fit_transform(scaled_x)
```

```python
pc_result= pca_model.fit_transform(scaled_x)
```

```python
plt.scatter(pc_result[:,0],pc_result[:,1])
plt.show()
```

```python
explained_variance=[]

for n in range(1,30):
    pca=PCA(n_components=n)
    pca.fit(scaled_x)
    
    explained_variance.append(np.sum(pca.explained_variance_ratio_))
```

```python
# 可以解釋幾成的資料
plt.plot(range(1,30),explained_variance)
plt.xlabel('Num of components')
plt.ylabel('Variance Explained')
plt.show()
```
